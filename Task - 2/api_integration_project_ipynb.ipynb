```python
import requests

# API URL
url = "https://jsonplaceholder.typicode.com/posts"

# Fetch data from API
response = requests.get(url)

# Check response
if response.status_code == 200:

    # Convert JSON response to Python list
    data = response.json()

    print("Successfully fetched data!")
    print("Total posts fetched:", len(data))

    print("\nFirst 5 posts:\n")

    # Display first 5 posts
    for post in data[:5]:

        print(f"Post {post['id']}:")
        print(f"ID: {post['id']}")
        print(f"Title: {post['title'][:50]}...")
        print(f"Body: {post['body'][:100]}...")
        print("-" * 20)

    # SEARCH / FILTER FEATURE
    search_id = int(input("\nEnter post ID to search (1-100): "))

    found = False

    for post in data:
        if post['id'] == search_id:

            print("\nPost Found!")
            print("---------------------")
            print("ID:", post['id'])
            print("Title:", post['title'])
            print("Body:", post['body'])

            found = True
            break

    if not found:
        print("Post not found!")

else:
    print(f"Error fetching data: {response.status_code}")
```
