```python
# ==========================================
# DATA ANALYSIS PROJECT USING PANDAS
# ==========================================

# Import libraries
import pandas as pd

# Load dataset from online source
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"

df = pd.read_csv(url)

# Dataset loaded message
print("Dataset Loaded Successfully!\n")

# First 5 rows
print("First 5 Rows:\n")
print(df.head())

# Dataset information
print("\nDataset Information:\n")
print(df.info())

# Descriptive statistics
print("\nDescriptive Statistics:\n")
print(df.describe())

# Missing values
print("\nMissing Values:\n")
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

print("\nDuplicates Removed Successfully!")

# Filter data
print("\nRows with rm greater than 6:\n")

filtered_data = df[df['rm'] > 6]

print(filtered_data.head())

# Group data
print("\nAverage Price by Age:\n")

grouped_data = df.groupby('age')['medv'].mean()

print(grouped_data.head())

# Top expensive houses
print("\nTop 5 Expensive Houses:\n")

top_houses = df.sort_values(by='medv', ascending=False)

print(top_houses.head())

# Save cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)

print("\nCleaned Dataset Saved Successfully!")

# Project completed
print("\nData Analysis Project Completed Successfully!")
```
